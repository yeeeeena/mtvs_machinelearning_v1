```python
import pandas as pd
import os
import glob

# CSV 파일들이 있는 디렉토리 경로
path = r'C:\Projects\pjt_002_머신러닝\데이터전체'  # 실제 경로로 변경해주세요
# 데이터전체데이터전체C:\Projects\pjt_002_머신러닝\데이터전체

# 모든 CSV 파일의 리스트를 가져옵니다
all_files = glob.glob(os.path.join(path, "*.csv"))

# 빈 리스트를 생성하여 각 CSV 파일의 DataFrame을 저장합니다
li = []

# 각 파일을 읽어 리스트에 추가합니다
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

# 모든 DataFrame을 연결합니다
frame = pd.concat(li, axis=0, ignore_index=True)

# 결과를 확인합니다
# print(frame)

# 필요하다면 결과를 새로운 CSV 파일로 저장할 수 있습니다
# frame.to_csv("combined_csv.csv", index=False, encoding='utf-8-sig')

frame.info()
```
